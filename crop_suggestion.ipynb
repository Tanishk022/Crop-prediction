```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']

plt.figure(figsize=(15, 25))

# Histogram + KDE for each numerical column
for i, col in enumerate(cols, 1):
    plt.subplot(7, 2, (i-1)*2 + 1)
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.title(f"Distribution of Soil {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

# Boxplot for each numerical column
    plt.subplot(7, 2, (i-1)*2 + 2)
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f"Boxplot of Soil {col}")
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# removing outlier form N and P

def remove_outliers_iqr(df, cols):
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df

# Remove outliers from only K and P columns
df = remove_outliers_iqr(df, ['K', 'P'])
```

```python
# removing outlier form N and P

# 1) P (Phosphorus)
# Phosphorus values ka distribution bahut skewed tha.
# Normal range 30–70 hota hai, but dataset me 120+ extreme values the.
# Yeh values sensor error / unrealistic spikes ho sakte hain.
# ML model ko distort karte hain — isliye remove kiya.

# 2) K (Potassium)
# Potassium mostly 15–40 range me hota hai.
# Dataset me kuch values 200+ the, jo real soil values me rare aur unrealistic hain.
# Yeh extreme outliers prediction ko heavily affect karte.
# Isliye IQR se remove kiya.

cols = ['N', 'P']

plt.figure(figsize=(10, 12))

# Histogram + KDE for each numerical column
for i, col in enumerate(cols, 1):
    plt.subplot(2, 2, (i-1)*2 + 1)
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.title(f"Distribution of Soil {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

# Boxplot for each numerical column
    plt.subplot(2, 2, (i-1)*2 + 2)
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f"Boxplot of Soil {col}")
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```
